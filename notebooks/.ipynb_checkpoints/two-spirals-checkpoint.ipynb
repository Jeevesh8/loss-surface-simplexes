# Setup Functions/Classes

```python
def twospirals(n_points, noise=.3, random_state=920):
    """
     Returns the two spirals dataset.
    """
    n = np.sqrt(np.random.rand(n_points,1)) * 600 * (2*np.pi)/360
    d1x = -1.5*np.cos(n)*n + np.random.randn(n_points,1) * noise
    d1y =  1.5*np.sin(n)*n + np.random.randn(n_points,1) * noise
    return (np.vstack((np.hstack((d1x,d1y)),np.hstack((-d1x,-d1y)))),
            np.hstack((np.zeros(n_points),np.ones(n_points))))
```

```python
X, Y = twospirals(500, noise=1.4)
train_x = torch.FloatTensor(X)
train_y = torch.FloatTensor(Y).unsqueeze(-1)

X, Y = twospirals(100, noise=1.4)
test_x = torch.FloatTensor(X)
test_y = torch.FloatTensor(Y).unsqueeze(-1)
```

```python
def trainer(model, train_x, train_y, reg_par=1e-5, niters=10,
           print_every=100, print_=False, n_samples=5):
    
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=1e-4)
    loss_func = torch.nn.BCEWithLogitsLoss()

    for step in range(niters):
        optimizer.zero_grad()
        vol = model.total_volume() + 1e-4
        acc_loss = 0
        for ii in range(n_samples):
            outputs = model(train_x)
            acc_loss += loss_func(outputs, train_y)
        print(vol.item())
        loss = acc_loss #- reg_par * (vol).log()
        if print_:
            if step % print_every == 0:
                print(acc_loss.item())
                print(vol.item())
                print("\n") 
            
        loss.backward()
        optimizer.step()
```

## Fix the Regularization Parameters

```python
total_verts = 4
reg_pars = [0.]
for ii in range(2, total_verts+1):
    fix_pts = [False]*ii
    start_vert = len(fix_pts)

    out_dim = 1
    arch_kwargs = {"in_dim":2, "hidden_size":25,
                         "activation":torch.nn.ReLU(), "bias":True}

    simplex_model = SimplexNet(out_dim, BasicSimplex, n_vert=start_vert,
                           fix_points=fix_pts,
                          architecture_kwargs=arch_kwargs)
#     simplex_model = simplex_model.cuda()
    log_vol = (volume_loss(simplex_model) + 1e-4).log()
    reg_pars.append(max(1./log_vol.item(), 1e-8))
```

```python
def complex_volume(model, ind):
    cdist = gpytorch.kernels.Kernel().covar_dist
    n_vert = len(model.simplicial_complex[ind])
    total_vert = model.n_vert
        
    mat = torch.ones(n_vert+1, n_vert+1) - torch.eye(n_vert + 1)
    
    ## compute distance between parameters ##
    n_par = int(sum([p.numel() for p in model.parameters()])/n_vert)
    par_vecs = torch.zeros(n_vert, n_par)

    for ii, vv in enumerate(model.simplicial_complex[ind]):
        par_vecs[ii, :] = utils.flatten([p for p in model.net.parameters()][vv::total_vert])
        
    dist_mat = cdist(par_vecs, par_vecs).pow(2)
    print(dist_mat)
    mat[:n_vert, :n_vert] = dist_mat
    print(mat)
    norm = (math.factorial(n_vert-1)**2) * (2. ** (n_vert-1))
    return torch.abs(torch.det(mat))
```

```python
def compute_loss_surface(model, train_x, train_y, v1, v2,
                        loss, n_pts=50, range_=10.):
    
    start_pars = model.state_dict()
    vec_lenx = torch.linspace(-range_.item(), range_.item(), 50)
    vec_leny = torch.linspace(-range_.item(), range_.item(), 40)
    ## init loss surface and the vector multipliers ##
    loss_surf = torch.zeros(50, 40)
    with torch.no_grad():
        ## loop and get loss at each point ##
        for ii in range(50):
            for jj in range(40):
                perturb = v1.mul(vec_lenx[ii]) + v2.mul(vec_leny[jj])
                # print(perturb.shape)
                perturb = utils.unflatten_like(perturb.t(), model.parameters())
                for i, par in enumerate(model.parameters()):
                    par.data = par.data + perturb[i].to(par.device)

                loss_surf[ii, jj] = loss(model(train_x), train_y)

                model.load_state_dict(start_pars)

    X, Y = np.meshgrid(vec_lenx, vec_leny)
    return X, Y, loss_surf
```

```python
X, Y, surf = compute_loss_surface(base_model, train_x, train_y,
                                  v1, v2, torch.nn.BCEWithLogitsLoss(),
                                 range_=range_)
```

## Compute Accuracies

```python
def compute_acc_base(model, xx, yy,
                    citerion=torch.nn.BCEWithLogitsLoss()):
    preds = model(xx).sigmoid().round()
    n_correct = torch.where(preds == yy)[0].numel()
    loss = criterion(preds, yy)
    return n_correct/yy.numel(), loss.item()

def compute_acc_ensemble(model, xx, yy, n_mdl=50,
                        citerion=torch.nn.BCEWithLogitsLoss()):
    preds = torch.zeros_like(model(xx))
    with torch.no_grad():
        for mdl in range(n_mdl):
            preds += model(xx)
        
        preds = preds.div(n_mdl).sigmoid().round()
    n_correct = torch.where(preds == yy)[0].numel()
    loss = criterion(preds, yy)
    return n_correct/yy.numel(), loss.item()
```

```python
accs = []
for vv in range(simplex_model.n_vert):
    simplex_model.export_base_parameters(base_model, vv)
    acc, lss = compute_acc_base(simplex_model, test_x, test_y)
    accs.append([vv, acc, lss])
acc_df = pd.DataFrame(accs)
```
